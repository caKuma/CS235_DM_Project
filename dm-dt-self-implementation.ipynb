```python
# Self-implement RegressionTree
# Decision Tree Node Class
class TreeNode:
    def __init__(self, depth):
        self.depth = depth # Depth of node
        self.feature_index = None # Feature value
        self.threshold = None # Split threshold
        self.left = None # Left child node
        self.right = None # Right child node
        self.value = None # Target value for prediction
```

```python
# Decision Tree Regressor Class
class DecisionTreeRegressorS:
    def __init__(self, max_depth=None, min_samples_split=2, min_samples_leaf=1):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        
    def fit(self, X, y):
        self.root = self.build_tree(X, y, depth=0)

    def build_tree(self, X, y, depth):
        n_samples, n_features = X.shape
        unique_values = np.unique(y)

        # Create a leaf node if conditions are met (if max_depth is had reached or min_sample_split is met)
        if (len(unique_values) == 1 or 
            (self.max_depth is not None and depth == self.max_depth) or 
            n_samples < self.min_samples_split):
            leaf = TreeNode(depth)
            leaf.value = np.mean(y)
            return leaf

        # Find the best split
        feature_index, threshold = self.find_best_split(X, y)

        # If no valid split was found, return a leaf node
        if feature_index is None:
            leaf = TreeNode(depth)
            leaf.value = np.mean(y)
            return leaf

        # Split the data
        X_left, y_left, X_right, y_right = self.split_data(X, y, feature_index, threshold)

        # Create a decision node
        decision_node = TreeNode(depth)
        decision_node.feature_index = feature_index
        decision_node.threshold = threshold
        decision_node.left = self.build_tree(X_left, y_left, depth + 1)
        decision_node.right = self.build_tree(X_right, y_right, depth + 1)

        return decision_node

    def find_best_split(self, X, y):
        n_samples, n_features = X.shape
        if n_samples <= 1:
            return None, None
        
        # Initialize variables
        best_mse = float('inf')
        best_feature_index = None
        best_threshold = None
        
        # Iterate through the feature and determine unique values
        for feature_index in range(n_features):
            thresholds = np.unique(X[:, feature_index])

            for threshold in thresholds:
                X_left, y_left, X_right, y_right = self.split_data(X, y, feature_index, threshold)

                # Check if the split respects the min_samples_leaf constraint
                if (len(y_left) < self.min_samples_leaf or 
                    len(y_right) < self.min_samples_leaf):
                    continue
                
                # Calculate / Minimize MSE for the split
                mse_left = np.var(y_left) * len(y_left)
                mse_right = np.var(y_right) * len(y_right)
                mse_split = mse_left + mse_right
                
                if mse_split < best_mse:
                    best_mse = mse_split
                    best_feature_index = feature_index
                    best_threshold = threshold
        
        # Returns feature value and split threshold
        return best_feature_index, best_threshold

    # Splits the data based on threshold
    def split_data(self, X, y, feature_index, threshold):
        left = X[:, feature_index] <= threshold
        right = ~left
        return X[left], y[left], X[right], y[right]
    
    # Runs predict_tree function for all nodes of from root to leaf node. 
    def predict(self, X):
        return np.array([self.predict_tree(x, self.root) for x in X])

    # Returns the leaf node's value for prediction.
    def predict_tree(self, x, node):
        if node.value is not None:
            return node.value

        if x[node.feature_index] <= node.threshold:
            return self.predict_tree(x, node.left)
        else:
            return self.predict_tree(x, node.right)
```

```python
# Regression Tree
# enc = preprocessing.LabelEncoder()
y = score['score']
X = train_scaled_t.to_numpy()
```

```python
# Model Fitting/Testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=40)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
model = DecisionTreeRegressorS(max_depth=5, min_samples_split=5, min_samples_leaf=20)
model.fit(X_train, y_train)
predicted = model.predict(X_test)
#accuracy = accuracy_score(predicted, y_test)
mae = mean_absolute_error(predicted, y_test)
rmse = mean_squared_error(predicted, y_test, squared=False)
mse = mean_squared_error(predicted, y_test)
#print(f'Accuracy: {accuracy}')
print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```

```python
#T-Test
#Equal Variance, hence equal_var = 'True'
tstat, t_pval = stats.ttest_ind(a=predicted, b=y_test, equal_var=True)

display("T-Statistic: " + str(tstat.round(3)), "PValue: " + str(t_pval.round(3)))
```
